```python
# Import AutoScraper
from autoscraper import AutoScraper
```

```python
scraper_url = "https://www.amazon.in/s?k=watch&crid=33XRV2UUQGIEJ&sprefix=watc%2Caps%2C271&ref=nb_sb_noss_2"
wanted_list = ["Multi Functional Sports Digital Multicolor Dial Men's Watch","₹299"]
```

```python
scraper = AutoScraper()
result = scraper.build(scraper_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(scraper_url, grouped=True)
```

```python
result = scraper.get_result_similar(scraper_url, grouped=True)
result
```

```python
result['rule_7naw'][0] 
```

```python
result['rule_5jql'][0]
```

```python
scraper.set_rule_aliases({'rule_7naw':'Title','rule_5jql':'Price'})
scraper.keep_rules(['rule_7naw','rule_5jql'])
scraper.save('amazon_watch')
```
